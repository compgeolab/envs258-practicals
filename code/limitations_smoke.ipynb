# Example of limitations of optical: smoke

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from functions import load_scene, rgb_composite
```

Load the pre-cropped scene.

```python
scene = load_scene(
    Path("../data/new_south_wales/LC08_L1TP_090083_20191215_20201022_02_T1_cropped/"),
    pattern="[2-4]",
)
scene_l2 = load_scene(
    Path("../data/new_south_wales/LC08_L2SP_090083_20191215_20201022_02_T1_cropped/"),
    pattern="[7]",
)
```

Plot an RBG and band 7 for comparison.

```python
rgb = rgb_composite(scene, in_range=(5000, 15000))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8.4), dpi=100)
ax1, ax2 = axes

ax1.imshow(rgb)

ax2.imshow(scene_l2[7], cmap="gist_heat")

for ax, title in zip(axes, ["RGB", "band 7 (SWIR)"]):
    ax.axis("off")      
    ax.text(
        25, 25, 
        title, 
        color="black", 
        fontweight="normal", 
        fontsize=14,
        verticalalignment="top",
        backgroundcolor="white",
    )    


for ax in axes:
    ax.axis("off")

fig.text(0, 0, 
    " Landsat 8 scene of New South Wales, Australia (2019/12/15)", 
    color="black", 
    fontweight="normal", 
    fontsize=11,
    verticalalignment="bottom",
    backgroundcolor="white",
)
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
plt.savefig("../lectures/images/new_south_wales_fire.svg", bbox_inches="tight", pad_inches=0)
```
